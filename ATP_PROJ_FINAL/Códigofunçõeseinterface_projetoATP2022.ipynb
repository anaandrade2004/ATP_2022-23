```python
import PySimpleGUI as sg
import os.path
import json 
import matplotlib.pyplot as plt

#LER DATASET
def lerdataset (basedados):
    f = open(basedados, encoding = "utf-8")
    bd = json.load(f)
    f.close()
    return bd

bd = lerdataset("dataset.json") #ler o dataset principal

#ADICIONAR DATASET
def adicionardataset(basedados):
    f1 = open(basedados, encoding='utf-8')
    data = json.load(f1)
    for pessoa in data['pessoas']:
        if pessoa not in bd['pessoas']:
            bd['pessoas'].append(pessoa)
    f1.close()
    return bd

#GRAVAR DATASET
def gravarBD(fnome):
    d=open(fnome,"w",encoding="utf-8")
    json.dump(bd,d)
    d.close()

#NÚMERO TOTAL DE INDIVÍDUOS
def totalindividuos(basedados):
    total = 0
    for pessoa in range(0, len(basedados["pessoas"])):
        total = total + 1
    return total 

#CARREGAR DATASET
def carregarbd():
    layout=[
        [sg.Text("Escreva o nome do dataset que pretende adicionar:")],
        [sg.Input(key="dataset")],
        [sg.Button("Ok"),sg.Button("Sair")]
    ]
    layout1=[
        [sg.Text("Informação guardada com sucesso.")]
    ]
    layout3= [
        [sg.Text("ERRO! O dataset que inseriu não existe.")]
    ]
    window= sg.Window("Adicionar dataset",layout)
    window1= sg.Window(" ",layout1)
    window2= sg.Window(" ",layout3)
    cond=True
    while cond:
        event,values=window.read()
        if event==sg.WIN_CLOSED or event=='Sair':
            cond=False
        elif event== "Ok":
            i=values["dataset"] 
            if os.path.isfile(i):
                adicionardataset(i)
                cond = True
                while cond:
                    event,values = window1.read()
                    if event == sg.WIN_CLOSED:
                        cond = False
                window.close()
            else:
                event,values = window2.read()
                if  event == sg.WIN_CLOSED:
                    cond = False
    window.close()

#GUARDAR DATASET
def guardar():
    layout=[
        [sg.Text("Escreva o nome que deseja para guardar o seu ficheiro:")],
        [sg.Input(key=("dataset"))],
        [sg.Button("Ok"),sg.Button("Sair")]
    ]
    layout1 = [
        [sg.Text("Dataset guardado com sucesso!")]
    ]
    layout2=[
        [sg.Text("ERRO! Não é possivel atribuir esse nome ao ficheiro.")]
    ]
    
    window=sg.Window("Guardar ficheiro",layout)
    window1= sg.Window("              ",layout1)
    window2 = sg.Window("              ",layout2)

    cond=True
    while cond:
        event,values=window.read()
        if event==sg.WIN_CLOSED or event=="Sair":
            cond=False
        elif event == "Ok":
            i = values["dataset"] 
            if values["dataset"] == "":
                cond = True                
                while cond:
                    event,values = window2.read()
                    if event == sg.WIN_CLOSED:
                        cond = False
                window.close()
            else:
                gravarBD(values["dataset"])
                cond = True
                while cond:
                    event,values = window1.read()
                    if event == sg.WIN_CLOSED:
                        cond = False
                window.close()
               
    window.close()  
    return i


#LISTAR POR NOME, RELIGIÃO, ANIMAIS, PESSOAS COM O MESMO ANIMAL PREFERIDO E PESSOAS COMA A MESMA RELIGIÃO
def listarporOrdNome(basedados):
    listaNomes = []
    for pessoa in basedados['pessoas']:
        if pessoa["nome"] not in listaNomes:
            listaNomes.append(pessoa['nome'])
    return sorted(listaNomes)

def listarReligiao(basedados):
    listaReligiao = []
    for pessoa in basedados['pessoas']:
        if "religiao" in pessoa.keys():
            if pessoa["religiao"] != "" and pessoa["religiao"] not in listaReligiao:
                listaReligiao.append(pessoa["religiao"])
    listaReligiao.sort()
    return listaReligiao

def listarAnimal(basedados):
    listaAnimais = []
    for pessoa in basedados['pessoas']:
        if "animais" in pessoa.keys():
            if pessoa["animais"]!= "":
                for i in pessoa["animais"]:
                    if i not in listaAnimais:
                        listaAnimais.append(i)
    listaAnimais.sort()
    return listaAnimais

def listarporanimalpreferido(basedados, animais):
    listaAnimais = []
    for pessoa in basedados['pessoas']:
        if "animais" in pessoa.keys():
            if animais in pessoa["animais"]:
                listaAnimais.append(pessoa["nome"])
    listaAnimais.sort()
    return listaAnimais

def listarmesmareligiao(basedados, religiao):
    listaReligiao = []
    for pessoa in basedados['pessoas']:
        if "religiao" in pessoa.keys():
            if religiao == pessoa["religiao"]:
                listaReligiao.append(pessoa["nome"])
    listaReligiao.sort()
    return listaReligiao

#JANELA DE INPUT DO ANIMAL E DA LISTA DE PESSOAS QUE GOSTA DESSE ANIMAL
def listarporAnimal():
    layout=[
        [sg.Text("Introduza o animal que deseja procurar:")],
        [sg.Input(key= "anim")],
        [sg.Button("Ok"),sg.Button("Sair")]
    ]
    window=sg.Window("Listar indivíduos por animal",layout)
    cond = True
    while cond:
        event,values = window.read()
        if event == sg.WIN_CLOSED or event == "Sair":
            cond = False
        elif event == "Ok":
            animal= values["anim"]
            listarpessoasanimal(bd,animal)
    window.close()

def listarpessoasanimal(bd,animal):
    listapessoas=listarporanimalpreferido(bd,animal)
    layout=[
        [sg.Table(listapessoas,["Lista de indivíduos com o mesmo animal"],justification="left")]
    ]
    window = sg.Window("",layout)
    cond = True
    while cond :
        event,values= window.read()
        if event == sg.WIN_CLOSED:
            cond = False
    window.close()

#JANELA DE INPUT DA RELIGIÃO E DA LISTA DE PESSOAS DESSA RELIGIÃO
def listarporReligiao():
    layout=[
        [sg.Text("Introduza a religião que deseja procurar:")],
        [sg.Input(key= "rel")],
        [sg.Button("Ok"),sg.Button("Sair")]
    ]
    window=sg.Window("Listar indivíduos por animal",layout)
    cond = True
    while cond:
        event,values = window.read()
        if event == sg.WIN_CLOSED or event == "Sair":
            cond = False
        elif event == "Ok":
            religiao= values["rel"]
            listarpessoasreligiao(bd,religiao)
    window.close()
    
def listarpessoasreligiao(bd,religiao):
    listapessoas=listarmesmareligiao(bd,religiao)
    layout=[
        [sg.Table(listapessoas,["Lista de indivíduos com a mesma religião"],justification="left")]
    ]
    window = sg.Window("",layout)
    cond = True
    while cond :
        event,values= window.read()
        if event == sg.WIN_CLOSED:
            cond = False
    window.close()

#JANELA PRINCIPAL DE INTERFACE PARA TODAS AS OPÇÕES DE ESCOLHA PARA LISTAGEM DE INDIVÍDUOS
def listar():
    listanomes = listarporOrdNome(bd)
    listareligioes = listarReligiao(bd)
    listaanimais = listarAnimal(bd)
    layout=[
        [sg.Text("Listar indivíduos por:")],
        [sg.Button("Ordem Alfabética")],
        [sg.Button("Religião")],
        [sg.Button("Animais")],
        [sg.Button("Indivíduos com o mesmo Animal")],
        [sg.Button("Indivíduos com a mesma Religião")],
        [sg.Button("Sair")]
    ]
    layout1 = [
        [sg.Table (listanomes,["Lista de nomes por ordem alfabética"],justification="left")], 
    ]
    layout2 = [
        [sg.Table (listareligioes,["Lista de religiões"],justification="left")],
        
    ]
    layout3 = [
        [sg.Table (listaanimais,["Lista de animais"],justification="left")],
        
    ]

    window=sg.Window("Listar indivíduos",layout)
    window1 = sg.Window("Listar por Ordem Alfabética",layout1)
    window2 = sg.Window("Listar por Religião",layout2)
    window3 = sg.Window("Listar por Animais",layout3)
    
    cond=True
    while cond:
        event,values=window.read()
        if event=="Sair" or event == sg.WIN_CLOSED:
            cond = False
        elif event == "Ordem Alfabética" :
            event1,values1 = window1.read()
            if event1 == sg.WIN_CLOSED:
                window1.close()
        elif event == "Religião":
            event2,values2 = window2.read()
            if event2 == sg.WIN_CLOSED:
                window2.close()
        elif event == "Animais":
            event3,values3 = window3.read()
            if event3 == sg.WIN_CLOSED:
                window3.close() 
        if event == "Indivíduos com o mesmo Animal":
            listarporAnimal()
        if event == "Indivíduos com a mesma Religião":
            listarporReligiao()
    window.close()

#INSERIR NOVO INDIVÍDUO
def novoindividuo():
    #sg.column(sg.text, ...., scrolable= True) - colocar toda a informação da página numa coluna para permitir utilizar a função scrollable uma vez que sem essa função, não era possivel visualizar a informação toda na tela e se colocassemos só scrollable no layout não iria reconhecer a função
    layout=[ 
        [sg.Column(([sg.Text("Nome:"), sg.Input(key =("nom"),do_not_clear = False)], #do_not_clear= False - elimina a informação para que o próximo usuário a introduzir nova informação não ncessite de apagar a informação anterior
        [sg.Text("Idade:"), sg.Slider(key=("id"), range=(0,100), orientation="horizontal")],
        [sg.Text("Sexo:"), sg.Input(key =("sex"),do_not_clear = False)],
        [sg.Text("Morada")],
        [sg.Text("Cidade:"), sg.Input(key =("cid"),do_not_clear = False),sg.Text("Distrito:"), sg.Input(key =("dist"),do_not_clear = False)],
        [sg.Text("CC:"), sg.Input(key =("cc"),do_not_clear = False), sg.Text("ou"), sg.Text("BI:"), sg.Input(key =("bi"),do_not_clear = False)],
        [sg.Text("Descrição:"), sg.Input(key =("desc"),do_not_clear = False)],
        [sg.Text("Profissão:"), sg.Input(key =("prof"),do_not_clear = False)],
        [sg.Text("Partido Político:")],
        [sg.Text("party_abbr:"), sg.Input(key =("abbr"),do_not_clear = False)],
        [sg.Text("party_name:"), sg.Input(key =("name"),do_not_clear = False)],
        [sg.Text("Religião:"), sg.Input(key =("rel"),do_not_clear = False)],
        [sg.Text("Desportos (se tiver mais de que um separe os por vírgulas):"), sg.Input(key =("desp"),do_not_clear = False)],
        [sg.Text("Animais (se tiver mais de que um separe os por vírgulas):"), sg.Input(key =("anim"),do_not_clear = False)],
        [sg.Text("Figura Pública Portuguesa:"), sg.Input(key =("figpt"),do_not_clear = False)],
        [sg.Text("Marca Carro:"), sg.Input(key =("marc"),do_not_clear = False)],
        [sg.Text("Destinos Favoritos (se tiver mais de que um separe os por vírgulas):"), sg.Input(key =("destfav"),do_not_clear = False)],
        [sg.Text("Atributos (se tiver mais de que um separe os por vírgulas):"), sg.Input(key =("atrib"),do_not_clear=False)],
        [sg.Checkbox("Fumador",key=("fum"))],
        [sg.Checkbox("Gosta de cinema",key=("cinem"))],
        [sg.Checkbox("Gosta de viajar",key=("viaj"))],
        [sg.Checkbox("Acorda cedo",key=("acord"))],
        [sg.Checkbox("Gosta de ler",key=("ler"))],
        [sg.Checkbox("Gosta de música",key=("music"))],
        [sg.Checkbox("Gosta de comer",key=("comer"))],
        [sg.Checkbox("Gosta de animais de estimação",key=("animest"))],
        [sg.Checkbox("Gosta de dançar",key =("dan"))],
        [sg.Text("Comida Favorita:"), sg.Input(key=("comida"),do_not_clear=False)],
        [sg.Button("Ok"), sg.Button("Sair")]), scrollable=True)]   
    ]
    layout1=[
        [sg.Text("Novo Indivíduo inserido com sucesso")]
    ]
    window=sg.Window("Adicionar ao dataset",layout)
    window1=sg.Window("                    ",layout1)
    cond=True
    while cond:
        event,values=window.read()
        if event==sg.WIN_CLOSED or event=="Sair": #event igual ao texto do botão para exercer a função de fechar a janela ou a função predefinida do PySimpleGUI do sg.WINDOW_CLOSED que permite fechar a janela através do x no canto superior direito 
            cond=False
        elif event=="Ok":
            nome=values["nom"]
            idade=int(values["id"])
            sexo=values["sex"]
            cidade=values['cid']
            distrito=values['dist']
            cc=values["cc"]
            bi=values["bi"]
            descricao=values["desc"]
            profissao=values["prof"]
            abbr=values['abbr']
            party=values['name']
            religiao=values['rel']
            desportos=values['desp']
            animais=values['anim']
            figura=values['figpt']
            marca=values['marc']
            destinos=values['destfav']
            fumador=values['fum']
            cinema=values["cinem"]
            viajar=values["viaj"]
            acordarcedo=values["acord"]
            ler=values["ler"]
            musica=values["music"]
            comer=values["comer"]
            animestimacao=values["animest"]
            danca=values["dan"]
            comidafav=values["comida"]
            cond = True 
            while cond:
                event,values = window1.read()
                if event == sg.WIN_CLOSED:
                    cond = False
            window1.close()

    listadesp=[]
    listaanim = []
    listafig=[]
    listadest=[]


    if "," in desportos:
        variosdesportos=desportos.split(",")
        for i in range(len(variosdesportos)):
            listadesp.append(variosdesportos[i])
    else:
        listadesp.append(desportos)                


    if "," in animais:
        variosanimais=animais.split(",")
        for i in range(len(variosanimais)):
            listaanim.append(variosanimais[i])
    else:
        listaanim.append(animais)      


    if "," in figura:
        variasfigura=figura.split(",")
        for i in range(len(variasfigura)):
            listafig.append(variasfigura[i])
    else:
        listafig.append(figura)     


    if "," in destinos:
        variosdestinos=destinos.split(",")
        for i in range(len(variosdestinos)):
            listadest.append(variosdestinos[i])
    else:
        listadest.append(destinos) 
    

    dicpartpol={}
    dicmorada={}
    dicatrib={}

    dicmorada["cidade"]=cidade
    dicmorada["distrito"]=distrito

    dicpartpol['party_abbr']=abbr
    dicpartpol['party_name']=party


    if fumador :
        dicatrib["fumador"] = True
    else:
        dicatrib["fumador"] = False

    if cinema:
        dicatrib["gosta_cinema"] = True
    else:
        dicatrib["gosta_cinema"] = False

    if viajar:
        dicatrib["gosta_viajar"] = True
    else:
        dicatrib["gosta_viajar"] = False

    if acordarcedo:
        dicatrib["acorda_cedo"] = True
    else:
        dicatrib["acorda_cedo"] = False
       
    if ler:
        dicatrib["gosta_ler"] = True
    else:
        dicatrib["gosta_ler"] = False

    if musica:
        dicatrib["gosta_musica"] = True
    else:
        dicatrib["gosta_musica"] = False

    if comer:
        dicatrib["gosta_comer"]=True
    else:
        dicatrib["gosta_comer"]=False
    
    if animestimacao:
        dicatrib["gosta_animais_estimacao"]=True
    else:
        dicatrib["gosta_animais_estimacao"]=False

    if danca:
        dicatrib["gosta_dancar"]=True
    else:
        dicatrib["gosta_dancar"]=False

    dicatrib["comida_favorita"] = comidafav



    dicionariofinal={}

    dicionariofinal["nome"]= nome
    dicionariofinal["idade"]= idade
    dicionariofinal["sexo"]= sexo
    dicionariofinal["morada"]= dicmorada
    if cc == "":
        dicionariofinal["BI"]= bi
    if bi == "":
        dicionariofinal["CC"]= cc
    dicionariofinal["descrição"]= descricao
    dicionariofinal["profissão"]= profissao
    dicionariofinal["partido_politico"]= dicpartpol
    dicionariofinal["religião"]=religiao
    dicionariofinal["desportos"]= listadesp
    dicionariofinal["animais"]=listaanim
    dicionariofinal["figura_publica_pt"]= figura
    dicionariofinal["marca_carro"]=marca
    dicionariofinal["destinos_favoritos"]=listadest
    dicionariofinal["atributos"]=dicatrib

    pessoa = dicionariofinal
    bd.update(pessoa)
    window.close()

#PROCURAR - PÁGINA INICIAL
def procurar():
    layout=[
        [sg.Text("Procurar por:")],
        [sg.Button("Nome")],
        [sg.Button("CC")],
        [sg.Button("BI")],
        [sg.Button("Sair")]
    ]

    window = sg.Window("Escolha uma opção para procurar um indivíduo",layout)
    i=0
    cond = True
    while cond:
        event, values = window.read()
        if event == "Sair" or event == sg.WIN_CLOSED:
            cond = False
        elif event == "Nome":
            procNome(i)
        elif event == "CC":
            procCC(i)
        elif event == "BI":
            procBI(i)
    window.close()

#PROCURAR POR NOME
def ProcurarNome(basedados, nome):
    i = []
    for pessoa in basedados['pessoas']:
        if nome in pessoa['nome'].split():
            i.append(pessoa['nome'])       
    return i 
    
def procurarpessoa(bd,nomec):
    i=[]
    for pessoa in bd['pessoas']:
        for info in pessoa:   
            if info=='nome' and nomec==pessoa['nome']:
                
                i=list(pessoa.items())
    return i

def procNome(i):
    layout = [
        [sg.Text("Introduza o nome da pessoa que deseja procurar:")],
        [sg.Input(key= "-PESSOA-")],
        [sg.Button("Ok"), sg.Button("Sair")]
    ]
    layout1=[
        [sg.Text("ERRO! Pessoa não existe.")]
    ]

    window = sg.Window("Procurar por Nome",layout)
    window1= sg.Window("ERRO",layout1)
    cond = True
    while cond:
        event, values = window.read()
        if event == "Sair" or event == sg.WIN_CLOSED:
            cond = False
        elif event == "Ok":
            if ProcurarNome(bd,values["-PESSOA-"]) == []:
                novocond = True
                while novocond:
                    event,values = window1.read()
                    if event == sg.WIN_CLOSED:
                        novocond = False
                window.close()
            else:
                nomep(values["-PESSOA-"])
    window.close()

def nomep(nomeinp):
    i=ProcurarNome(bd,nomeinp)

    layout = [
        [sg.Text("Escolha o nome da pessoa que deseja procurar:")],
        [sg.Combo(i,key="-PESSOAS-")],
        [sg.Button("Ok"), sg.Button("Sair")]
    ]

    window = sg.Window("Procurar por Nome",layout)

    cond = True
    while cond:
        event, values = window.read()
        if event == "Sair" or event == sg.WIN_CLOSED:
            cond = False
        elif event == "Ok":
            k =values ["-PESSOAS-"]
            mostrar_pessoaNOME(k)
    window.close()

def mostrar_pessoaNOME(k):
    pessoa=list(procurarpessoa(bd,k))
    nome=''
    idade=0
    sexo = ''
    id=''
    id1=''
    desc=''
    prof=''
    sigla=''
    extenso=''
    religião=''
    desp=''
    animais=''
    figPub=''
    carro=''
    destino=''
    cidade=''
    distrito=''
    atributos=''
    comida=''
    for info in pessoa:
            if info[0]== "nome":
                nome = info[1]
            elif info[0] == "idade":
                idade = info[1]
            elif info[0] == "sexo":
                sexo = info[1]
            elif info[0] in ["CC", "BI"]:
                id = info[1]
                id1 = info[0]
            elif info[0] == "descrição":
                desc = info[1]
            elif info[0] == "profissao":
                prof = info[1]
            elif info[0] == "partido_politico":
                sigla = info[1]["party_abbr"]
                extenso = info[1]["party_name"]
            elif info[0] == "religiao":
                religião = info[1]
            elif info[0] == "desportos":
                for i in info[1]:
                    desp= i +","+desp
            elif info[0] == "animais":
                for i in info[1]:
                    animais= i +","+ animais
            elif info[0] == "atributos":
                a = list(info[1].items())
                for tuplo in a:
                    if tuplo[0] == "comida_favorita":
                        comida=tuplo[1]
                    elif tuplo[0]=='fumador' and tuplo[1]:
                        fum= 'Fumador; '
                        atributos=atributos+fum
                    elif tuplo[0]=='gosta_cinema' and tuplo[1]:
                        cinema= 'Gosta de cinema; '
                        atributos=atributos+cinema
                    elif tuplo[0]=='gosta_viajar' and tuplo[1]:
                        viajar= 'Gosta de viajar; '
                        atributos=atributos+viajar
                    elif tuplo[0]=='acorda_cedo' and tuplo[1]:
                        cedo= 'Acorda cedo; '
                        atributos=atributos+cedo
                    elif tuplo[0]=='gosta_ler' and tuplo[1]:
                        ler= 'Gosta de ler; '
                        atributos=atributos+ler
                    elif tuplo[0]=='gosta_musica' and tuplo[1]:
                        musica= 'Gosta de musica; '
                        atributos=atributos+musica
                    elif tuplo[0]=='gosta_comer' and tuplo[1]:
                        comer= 'Gosta de comer; '
                        atributos=atributos+comer
                    elif tuplo[0]=='gosta_animais_estimacao' and tuplo[1]:
                        estimacao= 'Gosta de animais de estimação; '
                        atributos=atributos+estimacao
                    elif tuplo[0]=='gosta_dancar' and tuplo[1]:
                        dancar= 'Gosta de dançar; '
                        atributos=atributos+dancar
            elif info[0] == "figura_publica_pt":
                for i in info[1]:
                    figPub = i+'; ' + figPub
            elif info[0] == "marca_carro":
                carro = info[1]
            elif info[0] == "destinos_favoritos":
                for i in info[1]:
                    destino = i +'; '+ destino
            elif info[0] == "morada":
                cidade = info[1]["cidade"]
                distrito = info[1]["distrito"]
    layout= [
        [sg.Text("Nome"), sg.Text(nome)], 
        [sg.Text("Idade"), sg.Text(idade)],
        [sg.Text("Sexo"), sg.Text(sexo)],
        [sg.Text("Morada"), sg.Text("Cidade:"+ " " + cidade + " " + "Distrito:" + " " + distrito)],
        [sg.Text("CC/BI"), sg.Text(id1 + " : " + id)],
        [sg.Text("Descrição"), sg.Text(desc)],
        [sg.Text("Profissao"), sg.Text(prof)],
        [sg.Text("Partido_politico"), sg.Text(sigla + " - " + extenso)],
        [sg.Text("Religiao"), sg.Text(religião)],
        [sg.Text("Desportos"), sg.Text(desp)],
        [sg.Text("Animais"), sg.Text(animais)],
        [sg.Text("Figura publica"), sg.Text(figPub)],
        [sg.Text("Marca carro"), sg.Text(carro)],
        [sg.Text("Destinos_favoritos"), sg.Text(destino)],
        [sg.Text("Atributos:"), sg.Text(atributos)],
        [sg.Text('Comida Favorita:'),sg.Text(comida)],
        [sg.Button("Editar"),sg.Button("Sair")]
    ]
    window = sg.Window("Informações",layout,resizable=True)
    cond = True
    while cond:
        event,values = window.read()
        if event == "Sair" or event == sg.WIN_CLOSED:
            cond = False
        elif event == "Editar":
            alterar(k)
    window.close()

#PROCURAR POR CC
def procurarpessoacc(bd,cc):
    i=[]
    for pessoa in bd['pessoas']:  
        for info in pessoa:
            if 'CC' == info and cc == pessoa['CC']:
                i=list(pessoa.items())

    return i

def procCC(i):
    layout = [
        [sg.Text("Introduza o CC da pessoa que deseja procurar:")],
        [sg.Input(key= "-PESSOA-")],
        [sg.Button("Ok"), sg.Button("Sair")]
    ]
    layout1=[
        [sg.Text("ERRO! Pessoa não existe.")],
        [sg.Button("Inserir")]
    ]
    window = sg.Window("Procurar por CC",layout)
    window1= sg.Window("ERRO",layout1)

    cond = True
    while cond:
        event, values = window.read()
        if event == "Sair" or event == sg.WIN_CLOSED:
            cond = False
        if event == "Ok":
            if procurarpessoacc(bd,values["-PESSOA-"]) == []:
                    novocond = True
                    while novocond:
                        event,values = window1.read()
                        if event == sg.WIN_CLOSED:
                            novocond = False
                        elif event == "Inserir":
                            novoindividuo()
                    window.close()
            else:
                mostrar_pessoaCC(values["-PESSOA-"])
    window.close()

def mostrar_pessoaCC(k):
    pessoa= list(procurarpessoacc(bd,k))
    nome=''
    idade=0
    sexo = ''
    id=''
    id1=''
    desc=''
    prof=''
    sigla=''
    extenso=''
    religião=''
    desp=''
    animais=''
    figPub=''
    carro=''
    destino=''
    cidade=''
    distrito=''
    atributos=''
    comida=''
    
   
    for info in pessoa:
            if info[0]== "nome":
                nome = info[1]
            elif info[0] == "idade":
                idade = info[1]
            elif info[0] == "sexo":
                sexo = info[1]
            elif info[0] in ["CC", "BI"]:
                id = info[1]
                id1 = info[0]
            elif info[0] == "descrição":
                desc = info[1]
            elif info[0] == "profissao":
                prof = info[1]
            elif info[0] == "partido_politico":
                sigla = info[1]["party_abbr"]
                extenso = info[1]["party_name"]
            elif info[0] == "religiao":
                religião = info[1]
            elif info[0] == "desportos":
                for i in info[1]:
                    desp= i +","+desp
            elif info[0] == "animais":
                for i in info[1]:
                    animais= i +","+ animais
            elif info[0] == "atributos":
                a = list(info[1].items())
                for tuplo in a:
                    if tuplo[0] == "comida_favorita":
                        comida=tuplo[1]
                    elif tuplo[0]=='fumador' and tuplo[1]:
                        fum= 'Fumador; '
                        atributos=atributos+fum
                    elif tuplo[0]=='gosta_cinema' and tuplo[1]:
                        cinema= 'Gosta de cinema; '
                        atributos=atributos+cinema
                    elif tuplo[0]=='gosta_viajar' and tuplo[1]:
                        viajar= 'Gosta de viajar; '
                        atributos=atributos+viajar
                    elif tuplo[0]=='acorda_cedo' and tuplo[1]:
                        cedo= 'Acorda cedo; '
                        atributos=atributos+cedo
                    elif tuplo[0]=='gosta_ler' and tuplo[1]:
                        ler= 'Gosta de ler; '
                        atributos=atributos+ler
                    elif tuplo[0]=='gosta_musica' and tuplo[1]:
                        musica= 'Gosta de musica; '
                        atributos=atributos+musica
                    elif tuplo[0]=='gosta_comer' and tuplo[1]:
                        comer= 'Gosta de comer; '
                        atributos=atributos+comer
                    elif tuplo[0]=='gosta_animais_estimacao' and tuplo[1]:
                        estimacao= 'Gosta de animais de estimação; '
                        atributos=atributos+estimacao
                    elif tuplo[0]=='gosta_dancar' and tuplo[1]:
                        dancar= 'Gosta de dançar; '
                        atributos=atributos+dancar
            elif info[0] == "figura_publica_pt":
                for i in info[1]:
                    figPub = i+'; ' + figPub
            elif info[0] == "marca_carro":
                carro = info[1]
            elif info[0] == "destinos_favoritos":
                for i in info[1]:
                    destino = i +'; '+ destino
            elif info[0] == "morada":
                cidade = info[1]["cidade"]
                distrito = info[1]["distrito"]
    layout= [
        [sg.Text("Nome"), sg.Text(nome)], 
        [sg.Text("Idade"), sg.Text(idade)],
        [sg.Text("Sexo"), sg.Text(sexo)],
        [sg.Text("Morada"), sg.Text("Cidade:"+ " " + cidade + "       " + "Distrito:" + " " + distrito)],
        [sg.Text("CC/BI"), sg.Text(id1 + " : " + id)],
        [sg.Text("Descrição"), sg.Text(desc)],
        [sg.Text("Profissao"), sg.Text(prof)],
        [sg.Text("Partido_politico"), sg.Text(sigla + " - " + extenso)],
        [sg.Text("Religiao"), sg.Text(religião)],
        [sg.Text("Desportos"), sg.Text(desp)],
        [sg.Text("Animais"), sg.Text(animais)],
        [sg.Text("Figura publica"), sg.Text(figPub)],
        [sg.Text("Marca carro"), sg.Text(carro)],
        [sg.Text("Destinos_favoritos"), sg.Text(destino)],
        [sg.Text("Atributos:"), sg.Text(atributos)],
        [sg.Text('Comida Favorita:'),sg.Text(comida)],
        [sg.Button("Editar"),sg.Button("Sair")]
    ]

    window = sg.Window("Informações",layout,resizable=True)
    cond = True
    while cond:
        event,values = window.read()
        if event == "Sair" or event == sg.WIN_CLOSED:
            cond = False
        if event == "Editar":
            alterar(nome)
    window.close()

#PROCURAR POR BI
def procurarpessoabi(bd,bi):
    i=[]
    for pessoa in bd['pessoas']:  
        for info in pessoa:
            if 'BI' == info and bi == pessoa['BI']:
                i=list(pessoa.items())
    return i

def procBI(i):

    layout = [
        [sg.Text("Introduza o BI da pessoa que deseja procurar:")],
        [sg.Input(key= "-PESSOA-")],
        [sg.Button("Ok"), sg.Button("Sair")]
    ]
    layout1 = [
        [sg.Text("ERRO! Pessoa não existe.")],
        [sg.Button("Inserir")]
    ]

    window = sg.Window("Procurar por BI",layout)
    window1 = sg.Window("      ",layout1)
    cond = True
    while cond:
        event, values = window.read()
        if event == "Sair" or event == sg.WIN_CLOSED:
            cond = False
        if event == "Ok":
            if procurarpessoabi(bd,values["-PESSOA-"]) == []:
                novocond = True
                while novocond:
                    event,values = window1.read()
                    if event == sg.WIN_CLOSED:
                        novocond = False
                    elif event == "Inserir":
                        novoindividuo()
                window.close()
            else:
                mostrar_pessoaBI(values["-PESSOA-"])
    window.close()

def mostrar_pessoaBI(k):
    pessoa =procurarpessoabi(bd,k)

    nome=''
    idade=0
    sexo = ''
    id=''
    id1=''
    desc=''
    prof=''
    sigla=''
    extenso=''
    religião=''
    desp=''
    animais=''
    figPub=''
    carro=''
    destino=''
    cidade=''
    distrito=''
    atributos=''
    comida=''
    
    for info in pessoa:
            if info[0]== "nome":
                nome = info[1]
                print(nome)
            elif info[0] == "idade":
                idade = info[1]
            elif info[0] == "sexo":
                sexo = info[1]
            elif info[0] in ["CC", "BI"]:
                id = info[1]
                id1 = info[0]
            elif info[0] == "descrição":
                desc = info[1]
            elif info[0] == "profissao":
                prof = info[1]
            elif info[0] == "partido_politico":
                sigla = info[1]["party_abbr"]
                extenso = info[1]["party_name"]
            elif info[0] == "religiao":
                religião = info[1]
            elif info[0] == "desportos":
                for i in info[1]:
                    desp= i +","+desp
            elif info[0] == "animais":
                for i in info[1]:
                    animais= i +","+ animais
            elif info[0] == "atributos":
                a = list(info[1].items())
                for tuplo in a:
                    if tuplo[0] == "comida_favorita":
                        comida=tuplo[1]
                    elif tuplo[0]=='fumador' and tuplo[1]:
                        fum= 'Fumador; '
                        atributos=atributos+fum
                    elif tuplo[0]=='gosta_cinema' and tuplo[1]:
                        cinema= 'Gosta de cinema; '
                        atributos=atributos+cinema
                    elif tuplo[0]=='gosta_viajar' and tuplo[1]:
                        viajar= 'Gosta de viajar; '
                        atributos=atributos+viajar
                    elif tuplo[0]=='acorda_cedo' and tuplo[1]:
                        cedo= 'Acorda cedo; '
                        atributos=atributos+cedo
                    elif tuplo[0]=='gosta_ler' and tuplo[1]:
                        ler= 'Gosta de ler; '
                        atributos=atributos+ler
                    elif tuplo[0]=='gosta_musica' and tuplo[1]:
                        musica= 'Gosta de musica; '
                        atributos=atributos+musica
                    elif tuplo[0]=='gosta_comer' and tuplo[1]:
                        comer= 'Gosta de comer; '
                        atributos=atributos+comer
                    elif tuplo[0]=='gosta_animais_estimacao' and tuplo[1]:
                        estimacao= 'Gosta de animais de estimação; '
                        atributos=atributos+estimacao
                    elif tuplo[0]=='gosta_dancar' and tuplo[1]:
                        dancar= 'Gosta de dançar; '
                        atributos=atributos+dancar
            elif info[0] == "figura_publica_pt":
                for i in info[1]:
                    figPub = i+'; ' + figPub
            elif info[0] == "marca_carro":
                carro = info[1]
            elif info[0] == "destinos_favoritos":
                for i in info[1]:
                    destino = i +'; '+ destino
            elif info[0] == "morada":
                cidade = info[1]["cidade"]
                distrito = info[1]["distrito"]
    layout= [
        [sg.Text("Nome"), sg.Text(nome)], 
        [sg.Text("Idade"), sg.Text(idade)],
        [sg.Text("Sexo"), sg.Text(sexo)],
        [sg.Text("Morada"), sg.Text("Cidade:"+ " " + cidade + " " + "Distrito:" + " " + distrito)],
        [sg.Text("CC/BI"), sg.Text(id1 + " : " + id)],
        [sg.Text("Descrição"), sg.Text(desc)],
        [sg.Text("Profissao"), sg.Text(prof)],
        [sg.Text("Partido_politico"), sg.Text(sigla + " - " + extenso)],
        [sg.Text("Religiao"), sg.Text(religião)],
        [sg.Text("Desportos"), sg.Text(desp)],
        [sg.Text("Animais"), sg.Text(animais)],
        [sg.Text("Figura publica"), sg.Text(figPub)],
        [sg.Text("Marca carro"), sg.Text(carro)],
        [sg.Text("Destinos_favoritos"), sg.Text(destino)],
        [sg.Text("Atributos:"), sg.Text(atributos)],
        [sg.Text('Comida Favorita:'),sg.Text(comida)],
        [sg.Button("Editar"),sg.Button("Sair")]
    ]

    window = sg.Window("Informações",layout,resizable = True)
    cond = True
    while cond:
        event,values = window.read()
        if event == "Sair" or event == sg.WIN_CLOSED:
            cond = False
        elif event == "Editar":
            alterar(nome)
    window.close()

#DISTRIBUIÇÕES PROFISSÕES
def distProfissao(basedados):
    res = {}
    for pessoa in basedados['pessoas']:
        if "profissao" in pessoa.keys():
            if pessoa["profissao"] in res:
                res[pessoa["profissao"]] = res[pessoa["profissao"]] + 1
            else:
                res[pessoa["profissao"]] = 1
    return res

def dist10prof(basedados):
    res = distProfissao(basedados)
    dist10P = list(res.items())
    dist10P.sort(key=lambda x:x[1], reverse=True)
    lista = dist10P[0:10]
    dicio = dict(lista)
    return dicio

#GRÁFICOS DISTRIBUIÇÕES PROFISSÕES
def graficoProf():
    pessoas = list(bd)
    distribProf=dist10prof(bd)
    plt.figure()
    plt.title("Distribuição: Profissões dos indivíduos", fontsize = 12)
    plt.bar(distribProf.keys(), distribProf.values(), color = "green")
    plt.xticks([x for x in range (0, len(distribProf.keys()))], distribProf.keys(), rotation = "vertical")
    plt.yticks([y for y in range(0,(50),10)])
    plt.xlabel("Profissão", fontsize = 10)
    plt.ylabel("Número de indivíduos", fontsize = 10)
    plt.tight_layout()
    figura = plt.gcf()
    figura.set_size_inches(10, 5)
    figura.savefig("profissão.png")
    return "profissão.png"

def graficoprofissoes():
    grafico = graficoProf()
    layout = [
        [sg.Image(grafico)]
    ]
    window = sg.Window ("Distribuições",layout)
    cond = True
    while cond:
        event,values = window.read()
        if event == sg.WIN_CLOSED:
            cond = False
    window.close()

#DISTRIBUIÇÕES DESPORTO
def distDesportos(basedados):
    res = {}
    for pessoa in basedados['pessoas']:
        if "desportos" in pessoa:
            for desporto in pessoa["desportos"]:
                if desporto in res:
                    res[desporto] += 1
                else:
                    res[desporto] = 1
    return res

def dist10desp(basedados):
    res = distDesportos(basedados)
    dist10 = list(res.items())
    dist10.sort(key=lambda x:x[1], reverse=True)
    lista = dist10[0:10]
    dicio = dict(lista)
    return dicio

#GRÁFICO DISTRIBUIÇÕES DESPORTO
def graficoDesp():
    pessoas = list(bd)
    distribDesp = distDesportos(bd)
    valores = list(distribDesp.items())
    novaDistrib = dist10desp(bd)
    plt.figure()
    plt.title("Distribuição: Top 10 de Desportos Praticados", fontsize = 12)
    plt.bar(novaDistrib.keys(), novaDistrib.values(), color = "green", width = 0.3)
    plt.xticks([x for x in range (0, len(novaDistrib.keys()))], novaDistrib.keys(), rotation = "vertical")
    plt.yticks([y for y in range(0,(1500), 100)])
    plt.xlabel("Desporto", fontsize = 10)
    plt.ylabel("Número de indivíduos", fontsize = 10)
    plt.tight_layout()
    figura2 = plt.gcf()
    figura2.set_size_inches(10, 5)
    figura2.savefig("desporto.png")
    return "desporto.png"

def graficodesportos():
    grafico = graficoDesp()
    layout = [
        [sg.Image(grafico)]
    ]
    window = sg.Window ("Distribuições",layout,resizable= True)
    cond = True
    while cond:
        event,values = window.read()
        if event == sg.WIN_CLOSED:
            cond = False
    window.close()


#GRÁFICOS - PÁGINA INICIAL
def graficos():
    layout=[
        [sg.Text("Distribuições:")],
        [sg.Button("Top 10 Desportos")],
        [sg.Button("Top 10 Profissões")],
        [sg.Button ("Sair")]
    ]

    window = sg.Window("",layout)
    cond = True
    while cond:
        event,values= window.read()
        if event == "Sair" or event == sg.WIN_CLOSED:
            cond = False
        if event == "Top 10 Desportos":
            graficodesportos()
        if event == "Top 10 Profissões":
            graficoprofissoes()
    window.close()

#DISTRUBUIÇÃO POR PROFISSÃO-JANELA DE INTERFACE
def janelaprof():
    prof = distProfissao(bd)
    info = list(prof.items())
    layout= [
        [sg.Listbox(info, ["Distribuição por Profissão"],size = (60,15))]
    ]
    window = sg.Window("Distribuição de Profissões",layout)
    cond = True
    cond = True
    while cond:
        event,values = window.read()
        if event == sg.WIN_CLOSED:
            cond = False
    window.close()

#DISTRIBUIÇÃO TOP10 PROFISSÕES
def janelatop10profissoes():
    t10prof=dist10prof(bd)
    listat10prof = t10prof.items()
    layout = [
        [sg.Listbox(listat10prof,["Lista top10 profissões"],size = (65,12))]
    ]
    window = sg.Window("Distribuição top 10 Profissões",layout)
    cond = True
    while cond:
        event, values = window.read()
        if event == sg.WIN_CLOSED:
            cond = False
    window.close()

#DISTRUBUIÇÃO POR DESPORTO
def janeladesp():
    desp = distDesportos(bd)
    info = list(desp.items())
    layout= [
        [sg.Listbox(info, ["Distribuição por Desporto"],size = (60,15))]
    ]
    window = sg.Window("Distribuição de Desporto",layout)
    cond = True
    while cond:
        event,values = window.read()
        if event == sg.WIN_CLOSED:
            cond = False
    window.close()

#DISTRIBUIÇÃO TOP10 DESPORTOS
def janelatop10desportos():
    t10desp=dist10desp(bd)
    listat10desp=t10desp.items()
    layout = [
        [sg.Listbox(listat10desp,["Lista top10 desportos"],size = (60,12))]
    ]
    window = sg.Window("Distribuição top10 Desportos",layout)
    cond = True
    while cond:
        event, values = window.read()
        if event == sg.WIN_CLOSED:
            cond = False
    window.close()

#DISTRIBUIÇÕES JANELA DE INTERFACE PRINCIPAL (COM JANELA DE PROFISSÕES, DESPORTOS, TOP10 PROFISSÕES E TOP10 DESPORTOS)
def distribuicoes():
    layout=[
        [sg.Text("Distribuições")],
        [sg.Button("Distribuição por Profissões")],
        [sg.Button("Distribuição top10 Profissões")],
        [sg.Button("Distribuição por Desportos")],
        [sg.Button("Distribuição top10 Desportos")],
        [sg.Button("Sair")]
    ]
    
    window = sg.Window("",layout)
    cond = True
    while cond:
        event, values = window.read()
        if event== "Sair" or event == sg.WIN_CLOSED:
            cond = False
        elif event == "Distribuição por Profissões":
            janelaprof()
        elif event == "Distribuição top10 Profissões":
            janelatop10profissoes()
        elif event == "Distribuição por Desportos":
            janeladesp()
        elif event == "Distribuição top10 Desportos":
            janelatop10desportos()
    window.close()

#ALTERAR INFORMAÇÕES
def alterar(N):
    for pessoa in bd['pessoas']:
        if N==pessoa['nome'] :
            layout1=[
                [sg.Text("Selecione que pârametros pretende alterar:")] ,
                [sg.Radio("Nome",1,key=("nom"))],
                [sg.Radio("Idade",1,key=("id"))],
                [sg.Radio("Sexo",1,key=("sex"))],
                [sg.Radio("Cidade",1,key=("cid"))],
                [sg.Radio("Distrito",1,key=("dist"))],
                [sg.Radio("Descriçâo",1,key=("desc"))],
                [sg.Radio("Profissão",1,key=("prof"))],
                [sg.Radio("Partido Político",1,key =("partpol"))],
                [sg.Radio("Religião",1,key =("rel"))],
                [sg.Radio("Desportos",1,key =("desp"))],
                [sg.Radio("Animais",1,key =("anim"))],
                [sg.Radio("Figura Política Portuguesa",1,key =("figpt"))],
                [sg.Radio("Marca de Carro",1,key =("marc"))],
                [sg.Radio("Destinos Favoritos",1,key =("destfav"))],
                [sg.Button("Ok"), sg.Button("Sair")]
                 ]
      
               
            window=sg.Window("Escolha o que pretende alterar", layout1)  
          
           
            layout2=[
                [sg.Text("Adicione as novas informações:")], 
                [sg.Button("Ok"), sg.Button("Sair")]
            ]

            layout3= [
                [sg.Text("Edição efetuada com sucesso!")]
            ]
            window3= sg.Window("Alteração de Informações",layout3)

            cond=True
            while cond:
                event,values=window.read()
                if event==sg.WIN_CLOSED or event=='Sair':
                    cond=False
                elif event=='Ok':
                    layout2=[
                        [sg.Text("Adicione as novas informações:")],
                        [sg.Button("Ok"), sg.Button("Sair")]
                    ]
                    if values["nom"]:
                        layout2.append(tuple([sg.Text("Nome"), sg.Input(key=("novonome"),do_not_clear=False)]))
                        windown=sg.Window("Preencha o novo Nome",layout2)   
                        novacond=True
                        while novacond :
                            event,values=windown.read()
                            if event==sg.WIN_CLOSED or  event=="Sair":
                                novacond=False
                            elif event=="Ok":
                                novo=values["novonome"] 
                                pessoa["nome"]=novo 
                        
                                cond = True
                                while cond:
                                    event,values = window3.read()
                                    if event == sg.WIN_CLOSED:
                                        cond = False
                                window3.close()
                                    
                    elif values['id']==True:
                        layout2.append(tuple([sg.Text("Idade"), sg.Input(key=("novaidade"),do_not_clear=False)]))
                        windown=sg.Window("Preencha a nova idade",layout2)   
                        novacond=True
                        while novacond :
                            event,values=windown.read()
                            if event==sg.WIN_CLOSED or  event=="Sair":
                                novacond=False
                            elif event=="Ok":
                                novo=values["novaidade"] 
                                pessoa["idade"]=novo

                                cond = True
                                while cond:
                                    event,values = window3.read()
                                    if event == sg.WIN_CLOSED:
                                        cond = False
                                window3.close()

                    elif values["sex"]:
                        layout2.append(tuple([sg.Text("Sexo"), sg.Input(key="novosexo",do_not_clear=False)]))
                        windown=sg.Window("Preencha o novo sexo",layout2)   
                        novacond=True
                        while novacond :
                            event,values=windown.read()
                            if event==sg.WIN_CLOSED or  event=="Sair":
                                novacond=False
                            elif event=="Ok":
                                novo=values["novosexo"] 
                                pessoa["sexo"]=novo 

                                cond = True
                                while cond:
                                    event,values = window3.read()
                                    if event == sg.WIN_CLOSED:
                                        cond = False
                                window3.close()

                    elif values["cid"]:  
                        layout2.append([sg.Text("Cidade:"), sg.Input(key=("novacidade"), do_not_clear=False)])
                        windown=sg.Window("Preencha a nova cidade",layout2)   
                        novacond=True
                        while novacond:
                            event,values=windown.read()
                            if event==sg.WIN_CLOSED or  event=="Sair":
                                novacond=False
                            elif event=="Ok":
                                novo=values["novacidade"] 
                                pessoa["morada"]["cidade"]=novo

                                cond = True
                                while cond:
                                    event,values = window3.read()
                                    if event == sg.WIN_CLOSED:
                                        cond = False
                                window3.close()

                    elif values["dist"]:
                        layout2.append([sg.Text("Distrito:"), sg.Input(key=("novodistrito"), do_not_clear=False)])
                        windown=sg.Window("Preencha o novo distrito",layout2)   
                        novacond=True
                        while novacond:
                            event,values=windown.read()
                            if event==sg.WIN_CLOSED or  event=="Sair":
                                novacond=False
                            elif event=="Ok":
                                novo=values["novodistrito"] 
                                pessoa["morada"]["distrito"]=novo 

                                cond = True
                                while cond:
                                    event,values = window3.read()
                                    if event == sg.WIN_CLOSED:
                                        cond = False
                                window3.close()

                    elif values["desc"]:  
                        layout2.append([sg.Text("Descrição:"), sg.Input(key=("novadescricao"),do_not_clear=False)])
                        windown=sg.Window("Preencha a nova descrição",layout2)   
                        novacond=True
                        while novacond :
                            event,values=windown.read()
                            if event==sg.WIN_CLOSED or  event=="Sair":
                                novacond=False
                            elif event=="Ok":
                                novo=values["novadescricao"] 
                                pessoa["descrição"]=novo

                                cond = True
                                while cond:
                                    event,values = window3.read()
                                    if event == sg.WIN_CLOSED:
                                        cond = False
                                window3.close()

                    elif values['prof']:  
                        layout2.append([sg.Text("Profissão:"), sg.Input(key=("novaprof"),do_not_clear=False)]) 
                        windown=sg.Window("Preencha a nova profissão",layout2)   
                        novacond=True
                        while novacond :
                            event,values=windown.read()
                            if event==sg.WIN_CLOSED or  event=="Sair":
                                novacond=False
                            elif event=="Ok":
                                novo=values["novaprof"] 
                                pessoa["profissao"]=novo

                                cond = True
                                while cond:
                                    event,values = window3.read()
                                    if event == sg.WIN_CLOSED:
                                        cond = False
                                window3.close()

                    elif values["partpol"]:
                        layout2.append([sg.Text("party_abbr:"), sg.Input(key=("novoabbr"),do_not_clear=False), sg.Text("party_name:"), sg.Input(key=("novoname"),do_not_clear=False)])
                        windown=sg.Window("Preencha os novos parâmetros correspondentes ao partido político",layout2)   
                        novacond=True
                        while novacond:
                            event,values=windown.read()
                            if event==sg.WIN_CLOSED or  event=="Sair":
                                novacond=False
                            elif event=="Ok":
                                novo=values["novoabbr"] 
                                pessoa["partido_politico"]["party_abbr"]=novo
                                novo1=values["novoname"]
                                pessoa["partido_politico"]["party_name"]=novo1

                                cond = True
                                while cond:
                                    event,values = window3.read()
                                    if event == sg.WIN_CLOSED:
                                        cond = False
                                window3.close()

                    elif values["rel"]:
                        layout2.append([sg.Text("Religião:"), sg.Input(key=("novareligiao"),do_not_clear=False)])
                        windown=sg.Window("Preencha a nova religião",layout2)   
                        novacond=True
                        while novacond :
                            event,values=windown.read()
                            if event==sg.WIN_CLOSED or  event=="Sair":
                                novacond=False
                            elif event=="Ok":
                                novo=values["novareligiao"] 
                                pessoa["religiao"]=novo

                                cond = True
                                while cond:
                                    event,values = window3.read()
                                    if event == sg.WIN_CLOSED:
                                        cond = False
                                window3.close()

                    elif values["desp"]:
                        layout2.append([sg.Text("Desportos:"), sg.Input(key=("novodesp"),do_not_clear=False)])
                        windown=sg.Window("Preencha o/s novo/s desporto/s",layout2)   
                        novacond=True
                        while novacond :
                            event,values=windown.read()
                            if event==sg.WIN_CLOSED or  event=="Sair":
                                novacond=False
                            elif event=="Ok":
                                novo=values["novodesp"] 
                                if "," in novo:
                                    muitos=novo.split(",")
                                    pessoa["desportos"]=muitos

                                    cond = True
                                    while cond:
                                        event,values = window3.read()
                                        if event == sg.WIN_CLOSED:
                                            cond = False
                                    window3.close()

                                else: 
                                    pessoa["desportos"]=novo
                                    cond = True
                                    while cond:
                                        event,values = window3.read()
                                        if event == sg.WIN_CLOSED:
                                            cond = False
                                    window3.close()
                                
                    elif values["anim"]:
                        layout2.append([sg.Text("Animais"), sg.Input(key=("novoanim"),do_not_clear=False)])
                        windown=sg.Window("Preencha o/s novo/s animal",layout2)   
                        novacond=True
                        while novacond :
                            event,values=windown.read()
                            if event==sg.WIN_CLOSED or  event=="Sair":
                                novacond=False
                            elif event=="Ok":
                                novo=values["novoanim"] 
                                if "," in novo:
                                    muitos=novo.split(",")
                                    pessoa["animais"]=muitos

                                    cond = True
                                    while cond:
                                        event,values = window3.read()
                                        if event == sg.WIN_CLOSED:
                                            cond = False
                                    window3.close()
                                else: 
                                    pessoa["animais"]=novo

                                    cond = True
                                    while cond:
                                        event,values = window3.read()
                                        if event == sg.WIN_CLOSED:
                                            cond = False
                                    window3.close()
                    elif values["figpt"]:
                        layout2.append([sg.Text("Figura Pública Portuguesa"), sg.Input(key=("novafigpt"),do_not_clear=False)])
                        windown=sg.Window("Preencha",layout2)   
                        novacond=True
                        while novacond :
                            event,values=windown.read()
                            if event==sg.WIN_CLOSED or  event=="Sair":
                                novacond=False
                            elif event=="Ok":
                                novo=values["novafigpt"] 
                                if "," in novo:
                                    muitos=novo.split(",")
                                    pessoa['figura_publica_pt']=muitos 

                                    cond = True
                                    while cond:
                                        event,values = window3.read()
                                        if event == sg.WIN_CLOSED:
                                            cond = False
                                    window3.close()

                                else: 
                                    pessoa['figura_publica_pt']=novo

                                    cond = True
                                    while cond:
                                        event,values = window3.read()
                                        if event == sg.WIN_CLOSED:
                                            cond = False
                                    window3.close()

                    elif values["marc"]:
                        layout2.append([sg.Text("Marca do Carro"), sg.Input(key=("novamarca"),do_not_clear=False)])
                        windown=sg.Window("Preencha a nova marca do carro",layout2)   
                        novacond=True
                        while novacond :
                            event,values=windown.read()
                            if event==sg.WIN_CLOSED or  event=="Sair":
                                novacond=False
                            elif event=="Ok":
                                novo=values["novamarca"] 
                                pessoa["marca_carro"]=novo

                                cond = True
                                while cond:
                                    event,values = window3.read()
                                    if event == sg.WIN_CLOSED:
                                        cond = False
                                window3.close()

                    elif values["destfav"]:
                        layout2.append([sg.Text("Destinos Favoritos"), sg.Input(key=("novodestfav"),do_not_clear=False)])
                        windown=sg.Window("Preencha",layout2)   
                        novacond=True
                        while novacond :
                            event,values=windown.read()
                            if event==sg.WIN_CLOSED or  event=="Sair":
                                novacond=False
                            elif event=="Ok":
                                novo=values["novodestfav"] 
                                if "," in novo:
                                    muitos=novo.split(",")
                                    pessoa["destinos_favoritos"]=muitos

                                    cond = True
                                    while cond:
                                        event,values = window3.read()
                                        if event == sg.WIN_CLOSED:
                                            cond = False
                                    window3.close()

                                else: 
                                    pessoa["destinos_favoritos"]=novo
                                    cond = True
                                    while cond:
                                        event,values = window3.read()
                                        if event == sg.WIN_CLOSED:
                                            cond = False
                                    window3.close()
            
            window.close()

#JANELA PRINCIPAL - INTERFACE 
def janelaprincipal():
    layout= [
        [sg.Button("Carregar Dataset")],
        [sg.Button("Guardar Dataset")],
        [sg.Button("Número total de Indivíduos")],
        [sg.Button("Consultar Indivíduo")],
        [sg.Button("Inserir Indivíduo")],
        [sg.Button("Listar Indivíduo")],
        [sg.Button("Distribuições")],
        [sg.Button("Gráficos")],
        [sg.Button("Alterar Informações")],
        [sg.Button("Sair")]
    ]    
    window= sg.Window("Escolha uma opção", layout)
    cond = True
    while cond:
        event,values = window.read()
        if event == "Sair" or event == sg.WIN_CLOSED:
            cond = False
        elif event == "Carregar Dataset":
            carregarbd()
        elif event == "Guardar Dataset":
            guardar()
        elif event == "Número total de Indivíduos":
            layout1= [
                [sg.Text("O número total de indivíduos é:"+ str(totalindividuos(bd)))]
            ]
            window1 = sg.Window("Número de indivíduos",layout1)
            event1,values1 = window1.read()
            if event1 == sg.WIN_CLOSED:
                window1.close()
        elif event == "Inserir Indivíduo":
            procurar()
        elif event == "Consultar Indivíduo":
            procurar()
        elif event == "Listar Indivíduo":
            listar()
        elif event == "Distribuições":
            distribuicoes()
        elif event == "Gráficos":
            graficos()
        elif event == "Alterar Informações":
            procurar()
    window.close()

janelaprincipal()
```
